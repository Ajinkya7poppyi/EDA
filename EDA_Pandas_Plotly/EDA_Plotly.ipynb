**cells** – plotly.graph_objects.table.Cells instance or dict with compatible properties

  - An rgb/rgba string (e.g. 'rgb(255,0,0)')

#### Scatter Plot

The **'hoverlabel'** property is an instance of Hoverlabel
- A dict of string/value properties that will be passed to the Hoverlabel constructor. 
- Supported dict properties:
    - align - alignsrc - bgcolor - bgcolorsrc - bordercolor - bordercolorsrc - font - namelength - namelengthsrc

```python
fig = go.Figure(data=go.Scatter(x = df.radius_mean, 
                                y = df.texture_mean,
                                hoverinfo = ['x', 'y', 'text', 'name'],
                                mode = 'markers',
                                text = df.diagnosis,
                                name = 'hello',
                                hoverlabel = dict(bordercolor = 'darkslategray', bgcolor='lightsalmon', align = 'auto')))

fig.show()
```

#### Lines + Scatter Plot

    
Layout and Data Together

The **'mode'** property is a flaglist and may be specified as a string containing:
- Any combination of ['lines', 'markers', 'text'] joined with '+' characters (e.g. 'lines+markers') OR exactly one of ['none'] (e.g. 'none')

The **Figure** method 
Figure(data=None, layout=None, frames=None, skip_invalid=False, **kwargs)
- data:- The 'data' property is a tuple of trace instances that may be specified as:
    - A list or tuple of trace instances (e.g. [Scatter(...), Bar(...)])
    - A single trace instance (e.g. Scatter(...), Bar(...), etc.)
- layout:- The 'layout' property is an instance of Layout that may be specified as:
    - A dict of string/value properties that will be passed to the Layout constructor
- frames:- The 'frames' property is a tuple of instances of Frame that may be specified as:
    - A list or tuple of dicts of string/value properties that will be passed to the Frame constructor   

```python
# Creating Line Chart
traceline = go.Scatter(x = df.area_mean,
                    y = df.perimeter_mean,
                    mode = "lines+markers",
                    name = "Perimeter Mean",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= df.diagnosis)
# Creating Scatter Chart
tracescatter = go.Scatter(x = df.area_mean,
                    y = df.texture_mean,
                    mode = "markers",
                    name = "Texture Mean",
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text= df.diagnosis)
data = [traceline, tracescatter]
layout = dict(title = 'Permieter Mean and Texture Mean vs Area Mean',
              xaxis= dict(title= 'Area Mean',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Perimeter Mean',ticklen= 5,zeroline= False))
fig = go.Figure(data = data, layout = layout)
fig.show()
```

#### Bar Plot

The **'barmode'** property is an enumeration that may be specified as:

- One of the following enumeration values: ['stack', 'group', 'overlay', 'relative']

```python
# Prepare data drame
df_sample = df.iloc[:3,:]
# Creating Bar Chart
tracebar1 = go.Bar(x = df_sample.area_mean,
                y = df_sample.perimeter_mean,
                name = "Perimeter Mean",
                marker = dict(color = 'rgba(200, 200, 2, 0.8)',\
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text= df_sample.diagnosis)
# Creating Bar Chart
tracebar2 = go.Bar(x = df_sample.area_mean,
                y = df_sample.texture_mean,
                name = "Texture Mean",
                marker = dict(color = 'rgba(200, 80, 200, 0.8)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text= df_sample.diagnosis)

data = [tracebar1, tracebar2]
layout = go.Layout(barmode = 'group',
                   title= 'Diagnosis',
                   xaxis= dict(title='Area Mean'))
fig = go.Figure(data = data, layout = layout)
fig.show()
```

#### Box Plots
- Median (50th percentile) = middle value of the data set. Sort and take the data in the middle. It is also called 50% percentile that is 50% of data are less that median(50th quartile)(quartile)

- 25th percentile = quartile 1 (Q1) that is lower quartile

- 75th percentile = quartile 3 (Q3) that is higher quartile

- height of box = IQR = interquartile range = Q3-Q1

- Whiskers = 1.5 * IQR from the Q1 and Q3

- Outliers = being more than 1.5*IQR away from median commonly.



```python
# Creating Box Chart
tracebox1 = go.Box(y = df.radius_worst,
                    name = "Radius Worst",
                    marker=dict(color='rgb(50, 50, 160)'))
# Creating Box Chart
tracebox2 = go.Box(y = df.texture_worst,
                    name = "Texture Worst",
                    marker=dict(color='rgb(50, 190, 140)'))
data = [tracebox1, tracebox2]

layout = dict(title = 'Radious Texture Outlier Detection',
              yaxis= dict(title= 'Worst Values',ticklen= 5,zeroline= False))
fig = go.Figure(data = data, layout = layout)
fig.show()
```

#### Pairwise Correlation of columns

##### Method of correlation:

- pearson : standard correlation coefficient
  - The Pearson's correlation coefficient (r) is a measure of linear correlation between two variables. It's value lies between -1 and +1, -1 indicating total negative linear correlation, 0 indicating no linear correlation and 1 indicating total positive linear correlation. Furthermore, r is invariant under separate changes in location and scale of the two variables, implying that for a linear function the angle to the x-axis does not affect r.
  - To calculate r for two variables X and Y, one divides the covariance of X and Y by the product of their standard deviations.


- kendall : Kendall Tau correlation coefficient
    - Similarly to Spearman's rank correlation coefficient, the Kendall rank correlation coefficient (τ) measures ordinal association between two variables. It's value lies between -1 and +1, -1 indicating total negative correlation, 0 indicating no correlation and 1 indicating total positive correlation.
    - To calculate τ for two variables X and Y, one determines the number of concordant and discordant pairs of observations. τ is given by the number of concordant pairs minus the discordant pairs divided by the total number of pairs.


- spearman : Spearman rank correlation
    - The Spearman's rank correlation coefficient (ρ) is a measure of monotonic correlation between two variables, and is therefore better in catching nonlinear monotonic correlations than Pearson's r. It's value lies between -1 and +1, -1 indicating total negative monotonic correlation, 0 indicating no monotonic correlation and 1 indicating total positive monotonic correlation.
    - To calculate ρ for two variables X and Y, one divides the covariance of the rank variables of X and Y by the product of their standard deviations.

```python
df.corr(method = 'pearson')
```

#### Heatmap With Graph Objects

```python
z = df.corr(method = 'pearson')
z = z.iloc[0:15, 0:15]


# Creating Heatmap Chart
traceheatmap = go.Heatmap(z = np.round(z.values, 2),
                       x = list(z.index.values),
                       y = list(z.index.values),
                       hoverongaps = False,
                       colorscale = 'reds')
data = [traceheatmap]

layout = dict(title = 'Heatmap')

fig = go.Figure(data = data, layout = layout)
fig.show()
```

#### Heatmap With Figure Factory

```python
fig = ff.create_annotated_heatmap(z = np.round(z.values, 2),
                                  x = list(z.index.values),
                                  y = list(z.index.values),
                                  colorscale='Viridis')
fig.show()
```
